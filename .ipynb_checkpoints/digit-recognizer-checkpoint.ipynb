```python
import pandas as pd

train_df = pd.read_csv('/Users/mattiaruefenacht/Dropbox/Code/digit-recognition/digit-recognizer/train.csv')
test_df = pd.read_csv('/Users/mattiaruefenacht/Dropbox/Code/digit-recognition/digit-recognizer/test.csv')
sub_df = pd.read_csv('/Users/mattiaruefenacht/Dropbox/Code/digit-recognition/digit-recognizer/sample_submission.csv')
```

```python
train_df.isna().sum()
```

Dropping 'label' from input data and adding labels to output data.

```python
X_train = train_df.drop(columns=['label'],axis=1)
y_train = train_df['label']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
test_df
```

```python
X_test = scalar.fit_transform(test_df)
```

```python
sub_df
```

**Predictions using SVM**

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC()
```

```python
model_svc.fit(X_train, y_train)
```

```python
model_svc.score(X_train,y_train)
```

```python
y_predict = model_svc.predict(X_test)
```

```python
y_predict
```

```python
submission  = pd.DataFrame({'ImageId': sub_df.ImageId, 'label': y_predict})
```
